# Bivariate Analysis

```python
sns.scatterplot(x=df["Age"],y=df["Fare"])
plt.show()
```

# Numerical vs Catergorical


```python
for col1 in cat_cols:
    for col2 in num_cols:
        print("Box plot of ",col1 ,"vs ", col2)
        sns.boxplot(x=df[col1],y=df[col2])
        plt.show()
```

```python
for col1 in cat_cols:
    for col2 in num_cols:
        print("Bar plot of ",col1 ,"vs ", col2)
        sns.barplot(x=df[col1],y=df[col2])
        plt.show()
```

```python
sns.barplot(x=df["Pclass"],y=df["Survived"],ci=None)
plt.show()
```
